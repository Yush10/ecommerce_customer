```python
import pandas as pd
import numpy as np
```

```python
ec = pd.read_csv('ecommerce_customer_behavior.csv')
ec
```

```python
num_rows = ec.shape[0]
ec['ID'] = np.random.randint(1, 300, size=num_rows)
date_range = pd.date_range(start='2025-01-01', end='2025-07-31')

ec['Date_Accessed'] = np.random.choice(date_range, size=num_rows)
ec.to_csv('ecommerce_cust_behavior.csv', index=False)
```

```python
ec2 = pd.read_csv('ecommerce_cust_behavior.csv')
platform_map = {
    'Google': 1,
    'Facebook': 2,
    'Instagram': 3,
    'Direct': 4
}


ec2['Platform_Num'] = ec2['Referral'].map(platform_map)

ec2.to_csv('Cust_Behavior_Final.csv', index=False)

print(ec2.head())
```

```python
data = {
    'Platform': ['Google', 'Facebook', 'Instagram', 'Direct'],
    'Average CPC': [0.63, 0.70, 0.55, 1.00]
}
cpc = pd.DataFrame(data)
cpc.to_csv('cpc_table.csv', index=False)
print(cpc)
```
